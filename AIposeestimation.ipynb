#  Calculating Angles between the joints 

```python
# calculating the angle between any three points 
# in this case , we consider , leftshoulder , left elbow, left wrist 
def calculateangle(a,b,c) :
    a = np.array(a) 
    b = np.array(b)
    c = np.array(c) 
    
    radian = np.arctan2(c[1]-b[1] , c[0] - b[0] ) - np.arctan2(a[1] - b[1] , a[0] -b[0]) 
    angle = np.abs(radian * 180.0 / np.pi) 
    
    if angle > 180.0 :
        return 360 - angle 
    else :
        return angle 
```
